```python
# fill in certificate missing value
movies['Certificate'] = movies['Certificate'].fillna('No Certificate')

movies.isnull().sum()
```

```python
# fill missing value in meta score
movies.sort_values('IMDB_Rating', ascending = False)
#movies.head(10)

# meta score is review of user, thus use the mean of previous and after movie's Meta score after arrange according IMDB rating
movies['Meta_score'] = movies['Meta_score'].interpolate(method='linear')

# check is that all null value had been removed
movies.isnull().sum()
```

```python
print(movies['Gross'].dtype)

# Convert the 'Gross' column to string type if it's not already
if movies['Gross'].dtype != 'object':
    movies['Gross'] = movies['Gross'].astype(str)

# Now apply the replace function to remove commas
movies['Gross'] = movies['Gross'].str.replace(',', '')

# Convert the 'Gross' column to numeric, handling non-convertible values
movies['Gross'] = pd.to_numeric(movies['Gross'], errors='coerce')

# Calculate the mean of the 'Gross' column
mean_gross = movies['Gross'].mean()

# Fill the missing values in 'Gross' with the calculated mean
movies['Gross'] = movies['Gross'].fillna(mean_gross)

# Format the 'Gross' column to 2 decimal places
movies['Gross'] = movies['Gross'].apply(lambda x: f"{x:.2f}")

print(movies['Gross'].head())
```

```python
# Combine Star1, Star2, Star3, Star4 into a single column 'Star'
movies['Stars'] = movies[['Star1', 'Star2', 'Star3', 'Star4']].agg(', '.join, axis=1)

# Drop the original Star1, Star2, Star3, and Star4 columns
movies = movies.drop(columns=['Star1', 'Star2', 'Star3', 'Star4'])

# Verify the new 'Star' column
movies.head()
```

```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from nltk.stem import PorterStemmer

def chk_word(movies_infor):
    # Make all information into lowercase
    movies[movies_infor] = movies[movies_infor].str.lower()
    
    # Remove HTML tags (> <)
    movies[movies_infor] = movies[movies_infor].apply(lambda x: re.sub(r'<[^>]+>', '', x))
    
    # Remove punctuation (, . ! ?)
    movies[movies_infor] = movies[movies_infor].apply(lambda x: re.sub(r'[^\w\s]', '', x))
    
    # Remove numbers (1 2 3)
    movies[movies_infor] = movies[movies_infor].apply(lambda x: re.sub(r'\d+', '', x))
    
    # Remove stopwords (the an)
    movies[movies_infor] = movies[movies_infor].apply(lambda x: ' '.join([word for word in x.split() if word not in ENGLISH_STOP_WORDS]))
    
    # Apply stemming (running -> run)
    stemmer = PorterStemmer()
    movies[movies_infor] = movies[movies_infor].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))
    
    # Remove extra whitespaces
    movies[movies_infor] = movies[movies_infor].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
```

```python
chk_word("Genre")
chk_word("Overview")
chk_word("Director")
chk_word("Stars")
```
